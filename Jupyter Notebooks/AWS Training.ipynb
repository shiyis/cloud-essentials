```python
num = str(902.13)
if "." in num:
    bar = []
    foo = num.split('.')
    for f in foo:
        f = str(f)[::-1]
        bar.append(f)
else:
    bar = num[::-1]

foo = ""
for i in range(len(bar)):
    foo += bar[i] + "."
print(foo.strip('.'))
```

```python
memo = {0:0, 1:1}

def fibonacci(num):
    if num in memo:
        return memo[num]
    memo[num] = fibonacci(num - 1) + fibonacci(num - 2)    
    return memo[num]
```

```python
from math import sqrt

def prime_or_not(num):
    flag = 0
    if (num > 1):
        for i in range(2, int(sqrt(num))+1):
            if (num % i == 0):
                flag = 1
                break
        if flag == 0:
            print(f'{num} is prime.')
        else:
            print(f'{num} is not prime.')
    else:
        print(f'{num} is not prime.')
```
