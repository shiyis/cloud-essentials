### Question 1.

Create an SQL query that shows the TOP 3 authors who sold the most books in total!

```python
author_1 = ["book_1", "book_2"]
author_2 = ["book_3", "book_4", "book_5"]
author_3 = ["book_6"]
author1 = 'author_1'
author2 = 'author_2'
author3 = 'author_3'

authors_list = [(author1, author_1[0]), (author1, author_1[1]), (author2, author_2[0]), (author2, author_2[1]), (author2, author_2[2]), (author3, author_3[0])]
authors = pd.DataFrame(authors_list, columns=['author_name', 'book_name'])

print(authors)
```

```python
book_list = [('book_1', 1000), ('book_2', 1500), ('book_3', 34000), ('book_4', 29000), ('book_5', 40000), ('book_6', 4400)]
# books = {k:v for k, v in book_list}
books = pd.DataFrame(book_list, columns=['book_name', 'sold_copies'])
print(books)
```

```python
authors_and_books = authors.merge(books, how='inner', on='book_name')
top3 = authors_and_books.nlargest(3, 'sold_copies')
top3.head()
```

### Question 2.

Write an SQL query to find out how many users inserted more than 1000 but less than 2000 images in their presentations!

```python
table = [(7494212, 1535308430),
          (7494212, 1535308433),
          (1475185, 1535308444),
          (6946725, 1535308475),
          (6946725, 1535308476),
          (6946725, 1535308477)]

event_users = pd.DataFrame(table, columns = ['user_id', 'event_date_time'])
print(event_users)
```

```python
user_count = event_users['user_id'].value_counts().rename_axis('user_id').reset_index(name = 'counts')

# Edit boundaries to the desired value.
lower_boundary = 1
upper_boundary = 3

lower_threshold = user_count['counts'] > lower_boundary 
upper_threshold = user_count['counts'] < upper_boundary
user_count[lower_threshold & upper_threshold]
```

### Question 3.

Print every department where the average salary per employee is lower than $500!

```python
departments = [('Sales', 123, 'John Doe'),
('Sales', 211, 'Jane Smith'),
('HR', 556, 'Billy Bob'),
('Sales', 711, 'Robert Hayek'),
('Marketing', 235, 'Edward Jorgson'),
('Marketing', 236, 'Christine Packard')]

salaries = [(500, 123, 'John Doe'),
(600, 211, 'Jane Smith'),
(1000, 556, 'Billy Bob'),
(400, 711, 'Robert Hayek'),
(1200, 235, 'Edward Jorgson'),
(200, 236, 'Christine Packard')]

dept_df = pd.DataFrame(departments, columns=['department_name', 'employee_id', 'employee_name'])
salary_df = pd.DataFrame(salaries, columns=['salary', 'employee_id', 'employee_name'])
print(dept_df.head(), '\n\n', salary_df.head())
```

```python
dept_salary = dept_df.merge(salary_df, how='inner', on='employee_id')
dept_salary.columns
dept_salary = pd.DataFrame(dept_salary.groupby('department_name')['salary'].mean())

# Edit the threshold value to find the desired result.
dept_salary[dept_salary['salary']<=500]
```
